# 數據概覽

```python
import pandas
df = pandas.read_csv("database.csv")
df
```

# 將 quantity_bought_mouth 轉換為數值

```python
# 先觀察有幾種不同的數值
df["quantity_bought_mouth"].unique()

# 有 10 種數值
```

```python
# 採用正則表達式
import re, numpy

# 建立轉換函數
def quantity_bought_mouth_translater(var) : 
    if pandas.isna(var) :
        return 0
    else : 
        pass
    
    # 使用正则表达式查找匹配的数字
    pattern = r'(\d+)'
    matches = re.search(pattern, var)

    if matches is not None : 
        return matches.group(0)
    else : 
        return numpy.nan

df["quantity_bought_mouth_level"] = df["quantity_bought_mouth"].map(quantity_bought_mouth_translater)
df.head(3)
```

# 將 target_price 分為 貨幣單位、價格 兩欄位

## 貨幣單位

```python
def target_price_unit(var) : 
    if pandas.isna(var) :
        return numpy.nan
    else : 
        pass

    # 正则表达式模式，匹配 $ 之前的所有字符
    pattern = r'(.*)\$'
    matches = re.search(pattern, var)

    if matches is not None : 
        return matches.group(0).replace("$", "")
    else : 
        return numpy.nan

df["target_price_unit"] = df["target_price"].map(target_price_unit)
df.head(3)
```

## 價格

```python
def target_price_price(var) : 
    if pandas.isna(var) :
        return numpy.nan
    else : 
        pass
    
    # 正则表达式模式，匹配 $ 之後的所有字符
    pattern = r'(?<=\$)(.*)'
    matches = re.search(pattern, var)

    if matches is not None : 
        try : 
            return float(matches.group(0))
        except : 
            return numpy.nan
    else : 
        return numpy.nan

df["target_price_price"] = df["target_price"].map(target_price_price)
df.head(3)
```

# 將 target_star_text 分為 平均評價、最高評價 兩欄位	

## 平均評價

```python
def target_star_text_average(var) : 
    if pandas.isna(var) :
        return numpy.nan
    else : 
        pass

    # 正则表达式模式，匹配 $ 之前的所有字符
    pattern = r'\d+'
    matches = re.findall(pattern, var)

    if matches : 
        return float(matches[0] + "." + matches[1])
    else : 
        return numpy.nan

df["target_star_text_average"] = df["target_star_text"].map(target_star_text_average)
df.head(3)
```

## 最高評價

```python
def target_star_text_max(var) : 
    if pandas.isna(var) :
        return numpy.nan
    else : 
        pass

    # 正则表达式模式，匹配 $ 之前的所有字符
    pattern = r'\d+'
    matches = re.findall(pattern, var)

    if matches : 
        return float(matches[2])
    else : 
        return numpy.nan

df["target_star_text_max"] = df["target_star_text"].map(target_star_text_max)
df.head(3)
```

# 輸出資料

```python
df.to_csv("database_clean.csv", index= False)
```
